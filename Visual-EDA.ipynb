# Visual EDA Graphs of Reaction Times:

Below, I displayed a histogram of the reaction times with lines indicating the mean (in black) and median (in orange) for the reaction times of the data of the flanker experiment. I also displayed a boxplot of the data. These graphs display the same data that appeared in the computational EDA example.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading in the data:
df = pd.read_csv('s15.csv')

#Converting the reaction time to milliseconds
df.rt = df.rt * 1000
```

```python
#Creating a histogram to display the reaction time data:
df.rt.plot(kind='hist', title='Histogram of Reaction Times')
plt.xlabel('rt (ms)')
plt.axvline(df['rt'].median(), 0, 1, color='black', linestyle='-.')
plt.axvline(df['rt'].mean(), 0, 1, color='orange', linestyle='--')
plt.show

#Printing labels for the lines:
print('Median: Orange')
print('Mean: Black')
```

```python
#Creating a boxplot to display the reaction time data:
df.rt.plot(kind='box')
plt.title('Boxplot of Reaction Times')
plt.ylabel('rt (ms)')

plt.show()
```
